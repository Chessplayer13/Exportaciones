```python
def limpiar_valor(valor):
    valor = str(valor).strip()
    if ',' in valor:
        valor = valor.replace('.', '').replace(',', '.')
    return float(valor)

# Aplicamos la función
exp_24['FOBDOL'] = exp_24['FOBDOL'].apply(limpiar_valor)
exp_24['FOBDOL'].sum()
```

```python
df['POSAR6'] = df['POSAR'].str[:6]
df['POSAR6'].unique()
```

```python
agru= df.groupby(['AÑO','POSAR6'])['FOBDOL'].sum().reset_index().sort_values(['AÑO', 'FOBDOL'], ascending=[True, False])
```

```python
# Calcula el total de exportaciones por año
total_anual = agru.groupby('AÑO')['FOBDOL'].transform('sum')

# Agrega el porcentaje al DataFrame
agru['PORCENTAJE_ANUAL'] = round((agru['FOBDOL'] / total_anual) * 100,2)
```
